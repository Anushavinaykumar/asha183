```python
import matplotlib.pyplot as plt
plt.plot(df['Ename'], df['category'], marker='o', linestyle='-', color='b')
plt.xlabel('Ename')
plt.ylabel('age')
plt.title('employee table')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.hist(df['category'], bins=10, edgecolor='k')
plt.xlabel('salary')
plt.ylabel('Frequency')
plt.title('Histogram of employee table')
plt.show()
```
